```python
import pandas as pd
import time
import datetime as dt
```

```python
productos = pd.read_csv("Productos.csv")
productos
```

```python
#A
promedio_ventas_productos = productos.groupby("producto").agg({"ventas":"mean"})
promedio_ventas_productos.columns = ["Promedio de ventas"]
promedio_ventas_productos
```

```python
#B
min_max_prom_ventas = productos.groupby("producto").agg({"ventas":["min", "max", "mean"]})
min_max_prom_ventas.columns = ["Minimo ventas mensual", "Maximo ventas mensual", "Promedio ventas mensual"]
min_max_prom_ventas = min_max_prom_ventas.stack()
min_max_prom_ventas
```

```python
#C
min_max_prom_ventas = productos.groupby("producto").agg({"ventas":["min", "max", "mean"]})
nivel_0 = min_max_prom_ventas.columns.get_level_values(0)
nivel_1 = min_max_prom_ventas.columns.get_level_values(1)
min_max_prom_ventas.columns = nivel_0 + "_" + nivel_1
min_max_prom_ventas
```

```python
#D
min_max_prom_ventas = productos.groupby("producto").agg({"ventas":["min", "max", "mean"]})
nivel_0 = min_max_prom_ventas.columns.get_level_values(0)
nivel_1 = min_max_prom_ventas.columns.get_level_values(1)
min_max_prom_ventas.columns = nivel_0 + "_" + nivel_1
min_max_prom_ventas = min_max_prom_ventas.stack()
min_max_prom_ventas = min_max_prom_ventas.unstack(0)
min_max_prom_ventas
```

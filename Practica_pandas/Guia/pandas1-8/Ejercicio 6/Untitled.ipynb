```python
#6:
import pandas as pd
import numpy as np
df = pd.read_csv('Ejercicio 6 Guia 1 casos.csv')
df.head()
```

```python
df2 = pd.read_csv('Ejercicio 6 Guia 1 batman.csv')
df2.head()
```

```python
#A: Tasa de resolucion de casos de la fuerza policial por categoria de caso (solo casos donde no participo batman)
#Primero me quedo con los casos que no participo batman
dfA = df.set_index(df['id_caso'])
dfA = dfA.drop('id_caso', axis = 1)
dfA.head()
```

```python
dfA['respuesta batman'] = df2.set_index('id_caso')['respuesta']
dfA.head()
```

```python
dfA = dfA.fillna(0)
dfA.head()
```

```python
dfA = dfA.loc[dfA['respuesta batman'] == 0]
dfA.head()
```

```python
#Quito los casos abiertos porque seria injusto decir que un caso abierto no fue resuelto
dfA = dfA.loc[dfA['estado_caso'] != 1]
dfA.head()
```

```python
dfA['counter'] = 1
dfagrupado = dfA.groupby('estado_caso').agg('sum')
dfagrupado.head()
```

```python
dfagrupado['counter'].sum()
```

```python
dfagrupado.loc[2]
```

```python
porcionDeCasosResueltosSinBatman = dfagrupado.loc[2]['counter']/dfagrupado['counter'].sum()
porcionDeCasosResueltosSinBatman
```

```python
#B: Ahora la tasa de resolucion de casos donde batman respondio a la batiseñal
dfB = df.set_index(df['id_caso'])
dfB = dfB.drop('id_caso',axis = 1)
dfB['respuesta batman'] = df2.set_index('id_caso')['respuesta']
dfB = dfB.fillna(0)
dfB = dfB.loc[dfB['respuesta batman'] == 1]
dfB = dfB.loc[dfB['estado_caso'] != 1]
dfB['counter'] = 1
dfagrupado = dfB.groupby('estado_caso').agg('sum')
dfagrupado.head()
```

```python
porcionDeCasosResueltosConBatman = dfagrupado.loc[2]['counter']/dfagrupado['counter'].sum()
porcionDeCasosResueltosConBatman
```

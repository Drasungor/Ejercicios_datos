```python
locaciones = pd.read_csv('locacion.csv')
locaciones.sample(10)
```

```python
categorias = pd.read_csv('categorias_deportivas.csv')
categorias.head()
```

```python
categorias.info()
```

## Ejercicio a

```python
aux = eventos.merge(locaciones, left_on='id_locacion'\
                                     , right_on='id_locacion', how='outer')
aux.head()
```

```python
aux['fecha'] = pd.to_datetime(aux['fecha'])
aux_filtrado = aux[(aux['fecha'].dt.day==14)|(aux['fecha'].dt.day==15)]
aux_filtrado.head()
```

```python
eventos_por_locacion = aux_filtrado.groupby('nombre').agg({'cantidad_espectadores':'sum'})
eventos_por_locacion.sort_values(by='cantidad_espectadores', ascending=False, inplace=True)
eventos_por_locacion.head(1)
```

## Ejercicio b

```python
todo = aux.merge(categorias, left_on='id_categoria_deportiva', right_on='id_categoria_deportiva', how='outer')
todo.head()
```

```python
todo.rename(columns={'nombre_x':'nombre_sede', 'nombre_y':'categoria'}, inplace=True)
todo.head()
```

```python
eventos_excedidos = todo[todo['capacidad_extendida']<todo['cantidad_espectadores']]
eventos_excedidos.head()
```

```python
excedidos = list(zip(eventos_excedidos['nombre_evento'],eventos_excedidos['categoria']))
excedidos
```

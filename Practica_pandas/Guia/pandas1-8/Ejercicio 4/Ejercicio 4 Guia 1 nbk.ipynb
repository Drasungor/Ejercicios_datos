```python
#4
import pandas as pd
import numpy as np
df = pd.read_csv('Ejercicio 4 Guia 1 libros.csv')
df2 = pd.read_csv('Ejercicio 4 Guia 1 libros2.csv')
df.head()
```

```python
df2.head()
```

```python
#A: Obtener un dataframe con el TOP5 de libros segun rating
dfA = df.groupby('book_id').agg('mean')
dfA.head()
```

```python
dfA = dfA.drop('user_id',axis = 1)
dfA.head()
```

```python
dfA.nlargest(5,'rating')
```

```python
#B:Obtener un dataframe que indique que libros tienen una diferencia mayor al 20% entre los ratings del archivo 1 y 2
#Como los dataframe manejan un sistema de puntajes del 1 al 5, entiendo diferencia de 20% como diferencia de un punto
#Viendo los dataframe de practica, veo que deberian ser los libros 1, 3 y 5

dfB = dfA
dfB['rating2'] = df2.set_index('book_id')['avg_rating']
dfB.head()
```

```python
dfB = dfB.loc[abs(dfB['rating'] - dfB['rating2']) > 1]
dfB.head()
```

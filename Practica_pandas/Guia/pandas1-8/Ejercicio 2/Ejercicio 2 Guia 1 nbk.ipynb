```python
import datetime as datetime
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Ejercicio 2 Guia 1 transacciones.csv")
df.head()
```

```python
#Veo los tipos del df
df.dtypes
```

```python
#Cambio las fechas y la hora al formato apropiado
```

```python
df['fecha'] = pd.to_datetime(df['fecha'])
df['hora'] = pd.to_datetime(df['hora'])
df.head()
```

```python
#A: Validar que todas las transacciones cuenten con un tipo de transaccion
#Supondre que solo son validos los tipos de transaccion extraccion transferencia y deposito.
#Siendo los vacios un tipo de transaccion no valida.
df['tipo'].isna().sum()
```

```python
df['isna'] = df['tipo'].isna()
df.head()
```

```python
df = df.loc[~df['isna']]
```

```python
df.shape
```

```python
#Acá limpio la columna utilitaria que utilice
df = df.drop(['isna'], axis = 1)
df.head()
```

```python
#B: Verificar que para todas las transacciones exista tanto cuenta de origen como cuenta de destino
#Primero me armo dps columnas utilitarias para luego filtrar las filas de interes
#Quiero dropear las filas que, si la transaccion es de tipo transferencia, exista tanto cuenta de origen, como de destino
```

```python
df['cuenta origen isna'] = df['cuenta origen'].isna()
df['cuenta destino isna'] = df['cuenta destino'].isna()
df.head()
```

```python
dfFiltrado = df.loc[lambda row: ~((row['tipo'] == 'transferencia') & (row['cuenta origen isna'] | row['cuenta destino isna'])),:]
dfFiltrado.shape
```

```python
df.shape
```

```python
df = dfFiltrado
df.shape
```

```python
df= df.drop(['cuenta origen isna','cuenta destino isna'], axis = 1)
df.head()
```

```python
#C: Verificar que todas las transferencias tegan montos mayores a cero
```

```python
df.shape
```

```python
df = df.loc[df['monto'] > 0]
df.shape
```

```python
#D: Indicar cuales fueron las 10 transacciones de mayor monto
```

```python
df.nlargest(10,'monto')
```

```python
#E: Indicar el tipo de transaccion que registra el mayor promedio de monto.
```

```python
dfAgrupado = df.groupby('tipo').agg('mean')
dfAgrupado.nlargest(1,'monto')
```

```python
dfAgrupado['monto'].idxmax()
```

```python
#Ahí se muestra que los depósitos tienen el mayor promedio de montos.
```

```python
#F: Indicar cuales son las 5 cuentas con mayor cantidad de transacciones.
```

```python
df['counter'] = 1
df['counter2'] = 1
df.head()
```

```python
dfAgrupado = df.groupby('cuenta origen')['counter'].agg('sum')
#La idea es nunca imprimir el df entero, pero aqui lo hago para verificar que el resultado es el correcto
dfAgrupado
```

```python
#Esa es la cantidad de transacciones que tiene como cuenta ORIGEN a la cuenta objetivo
#Ahora tengo que involucrar a las que tienen como cuenta DESTINO
dfAgrupado2 = df.groupby('cuenta destino')['counter'].agg('sum')
dfAgrupado2
```

```python
dfFinal = dfAgrupado + dfAgrupado2
dfFinal
```

```python
dfFinal.nlargest(5)
```

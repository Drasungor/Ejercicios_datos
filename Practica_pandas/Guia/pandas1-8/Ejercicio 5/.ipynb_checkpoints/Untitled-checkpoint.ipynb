```python
import pandas as pd

hospitales = pd.read_csv('Ejercicio 5 Guia 1 hospitales.csv')
hospitales.head()
```

```python
hospitales.info()
```

```python
nacimientos = pd.read_csv('Ejercicio 5 Guia 1 nacimientos.csv')
nacimientos.head()
```

```python
nacimientos.info()
```

## Ejercicio a

```python
merge = hospitales.merge(nacimientos, left_on='id_hospital', right_on='id_hospital', how='outer')
merge.head()
```

```python
nacimientos_oct_2017 = merge[(merge['mes_nacimiento']==10) & (merge['anio_nacimiento']==2017)]
nacimientos_oct_2017.head()
```

```python
nacimientos_oct_2017_final = nacimientos_oct_2017.groupby('id_hospital')\
                        .agg({'promedio_nacimientos_mensual':['count','mean']})
nacimientos_oct_2017_final
```

```python
nivel0 = nacimientos_oct_2017_final.columns.get_level_values(0)
nivel1 = nacimientos_oct_2017_final.columns.get_level_values(1)
nacimientos_oct_2017_final.columns = nivel0 + '_' + nivel1
nacimientos_oct_2017_final.rename(columns={'promedio_nacimientos_mensual_count':'nacimientos', \
                                     'promedio_nacimientos_mensual_mean':'promedio_mensual'}, inplace=True)
nacimientos_oct_2017_final
```

## Ejercicio b

```python
nacimientos_oct_2016 = merge[(merge['mes_nacimiento']==10) & (merge['anio_nacimiento']==2016)]
nacimientos_oct_2016.head()
```

```python
oct_2016_2017 = pd.concat([nacimientos_oct_2016,nacimientos_oct_2017])
oct_2016_2017
```

```python
def por_cesarea(x):
    partos_cesarea = (x==2).sum()
    return (partos_cesarea/x.size)*100

oct_2016_2017 = oct_2016_2017.groupby('anio_nacimiento').agg({'tipo_parto':por_cesarea})
oct_2016_2017.rename(columns={'tipo_parto':'porcentaje_partos_cesarea'}, inplace=True)
oct_2016_2017
```

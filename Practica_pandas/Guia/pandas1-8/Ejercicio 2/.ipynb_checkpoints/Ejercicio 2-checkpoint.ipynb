```python
import pandas as pd
import numpy as np

transacciones = pd.read_csv('Ejercicio 2 Guia 1 transacciones.csv')
transacciones.head(50)
```

```python
transacciones.info()
```

```python
transacciones.set_index('id', inplace=True)
transacciones.head()
```

## Ejercicio a

```python
transacciones['tipo'].hasnans
```

```python
transacciones = transacciones[pd.notnull(transacciones['tipo'])]
transacciones['tipo'].hasnans
```

## Ejercicio b

```python
transacciones[transacciones['tipo'] == 'transferencia']['cuenta origen'].hasnans
```

```python
transacciones[transacciones['tipo'] == 'transferencia']['cuenta destino'].hasnans
```

```python
filas_nan = transacciones[(transacciones['tipo']=='transferencia') & \
                                    ((transacciones['cuenta origen'].isnull()) | (transacciones['cuenta destino'].isnull()))].index.to_list()
transacciones.drop(index=filas_nan, inplace=True)
```

```python
transacciones[transacciones['tipo'] == 'transferencia']['cuenta origen'].hasnans
```

```python
transacciones[transacciones['tipo'] == 'transferencia']['cuenta destino'].hasnans
```

## Ejercicio c

```python
(transacciones[transacciones['tipo'] == 'transferencia']['monto']==0).sum()
```

```python
(transacciones[transacciones['tipo'] == 'extraccion']['monto']==0).sum()
```

```python
(transacciones[transacciones['tipo'] == 'deposito']['monto']==0).sum()
```

```python
filas_nan = transacciones[transacciones['monto']==0 & ((transacciones['tipo']=='transferencia') | (transacciones['tipo']=='deposito') | (transacciones['tipo']=='extraccion'))].index.to_list()
transacciones.drop(index=filas_nan, inplace=True)
```

```python
(transacciones[transacciones['tipo'] == 'transferencia']['monto']==0).sum()
```

## Ejercicio d

```python
transacciones.sort_values(by='monto', ascending=False).head(10)
```

## Ejercicio e

```python
transacciones_por_tipo = transacciones.groupby('tipo').agg({'monto':'mean'}).sort_values(by='monto', ascending=False)
transacciones_por_tipo
```

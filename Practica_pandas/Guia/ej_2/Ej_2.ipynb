```python
import pandas as pd
import datetime as dt
import time
```

```python
tiempo_inicial = time.time()
transacciones = pd.read_csv("Transacciones.csv")
transacciones
```

```python
#A
transacciones_con_tipo = transacciones[~transacciones["Tipo de transaccion"].isnull()]
transacciones_con_tipo
```

```python
#B
transacciones_validas = transacciones[(transacciones["Tipo de transaccion"] != "Transferencia") |\
                                      ((~transacciones["Cuenta origen"].isnull()) & \
                                       (~transacciones["Cuenta destino"].isnull()))]
transacciones_validas
```

```python
#D
transacciones.nlargest(10, ["Monto"], "all")
```

```python
#E

promedio_monto_tipo = transacciones.groupby("Tipo de transaccion").agg({"Monto": "mean"})
promedio_monto_tipo.columns = ["promedio_monto"]
promedio_monto_tipo.nlargest(1, "promedio_monto", "all")
```

```python
#F
serie_cuenta_origen = transacciones.loc[:, "Cuenta origen"]
serie_cuenta_destino = transacciones.loc[:, "Cuenta destino"]
#serie_cuenta_origen = serie_cuenta_origen.dropna()
#serie_cuenta_destino = serie_cuenta_destino.dropna()
serie_cuenta_origen = serie_cuenta_origen.value_counts()
serie_cuenta_destino = serie_cuenta_destino.value_counts()
cantidad_total_cuenta = serie_cuenta_origen.add(serie_cuenta_destino, fill_value = 0)
cantidad_total_cuenta.nlargest(5, "all")
```

```python
#G
transacciones_por_cuenta_origen = transacciones.groupby("Cuenta origen")
transacciones_por_cuenta_destino = transacciones.groupby("Cuenta destino")
sumas_origen = transacciones_por_cuenta_origen["Monto"].sum()
sumas_destino = transacciones_por_cuenta_destino["Monto"].sum()
monto_total_por_cuenta = sumas_origen.add(sumas_destino, fill_value=0)
monto_total_por_cuenta.nlargest(5, "all")
```

```python
#H
#Lo ideal seria primero guardar el groupby y despues hacer el .agg, para despues agarrar el grupo 
#de la transaccion que se quiere sin tener que volver a filtrar
transacciones_por_tipo = transacciones.groupby("Tipo de transaccion").agg({"Monto": "mean"})
transaccion_mayor_promedio = transacciones_por_tipo.nlargest(1, "Monto")
transaccion_mayor_promedio = transaccion_mayor_promedio.index[0]
transacciones_mayor_promedio = transacciones[transacciones["Tipo de transaccion"] == transaccion_mayor_promedio]
transacciones_origen = transacciones_mayor_promedio["Cuenta origen"].value_counts()
transacciones_destino = transacciones_mayor_promedio["Cuenta destino"].value_counts()
transacciones_totales_por_cuenta = transacciones_origen.add(transacciones_destino, fill_value = 0)
transacciones_totales_por_cuenta.nlargest(5, "all")
```

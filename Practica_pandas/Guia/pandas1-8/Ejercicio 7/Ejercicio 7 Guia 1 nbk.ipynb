```python
#7: Analizar los dataframe de los juegos olimpicos que tanto me costo armar. Dios.
import pandas as pd
import numpy as np
eventos = pd.read_csv('eventos.csv')
locaciones = pd.read_csv('locacion.csv')
catdep = pd.read_csv('categorias_deportivas.csv')
eventos.head()
```

```python
locaciones.head()
```

```python
catdep.head()
```

```python
#A: Nombre de la sede que acumulo la mayor cantidad de espectadores entre el 14 y el 15 de octubre
eventos['dia'] = pd.to_datetime(eventos['fecha']).dt.day
eventos.head()
```

```python
dfA = eventos.loc[(eventos['dia'] == 14) | (eventos['dia'] == 15)]
dfA.head()
```

```python
dfA = dfA.groupby('id_locacion').agg('sum')
dfA.head()
```

```python
dfA = dfA.drop(['id_evento','id_categoria_deportiva','dia'], axis = 1)
dfA.head()
```

```python
locacionesA = locaciones.set_index('id_locacion')
locacionesA.head()
```

```python
locacionesA['espectadores'] = dfA['cantidad_espectadores']
locacionesA.head()
```

```python
locacionesA = locacionesA.fillna(0)
locacionesA = locacionesA.groupby('sede').agg('sum')
locacionesA.head()
```

```python
locacionesA.nlargest(4,'espectadores')
```

```python
#B: Nombre del evento y nombre de la categoria deportiva de aquellos eventos cuya cantidad de espectadores supera la capacidad extendida de la locacion donde tuvo curso el evento
dfB = eventos.set_index('id_locacion')
dfB.head()
```

```python
dfB['capacidad_extendida'] = locaciones.set_index('id_locacion')['capacidad_extendida']
dfB.head()
```

```python
dfB = dfB.reset_index()
dfB = dfB.set_index('id_categoria_deportiva')
dfB['nombre_categoria_deportiva'] = catdep.set_index('id_categoria_deportiva')['nombre']
dfB.head()
```

```python
dfB = dfB.reset_index()
dfB = dfB.loc[dfB['cantidad_espectadores'] > dfB['capacidad_extendida']]
dfB.head()
```

```python
dfFinal = dfB.drop(['id_categoria_deportiva','id_locacion','id_evento','fecha','cantidad_espectadores','dia','capacidad_extendida'], axis = 1)
dfFinal.head()
```

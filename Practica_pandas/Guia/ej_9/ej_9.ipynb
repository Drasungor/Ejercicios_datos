```python
import pandas as pd
import datetime as dt
```

```python
internal = pd.read_csv("Internal.csv")
scrapping = pd.read_csv("Scrapping.csv")
```

```python
#A
operaciones_internal = internal.groupby("vehicle_model_id").agg({"price":["mean","std"]})
operaciones_internal.columns = ["int_mean_price", "int_std_price"]
operaciones_scrapping = scrapping.groupby("vehicle_model_id").agg({"price":["mean","std"]})
operaciones_scrapping.columns = ["ext_mean_price", "ext_std_price"]
df_final = operaciones_scrapping.merge(operaciones_internal, how="outer", on="vehicle_model_id")
df_final
```

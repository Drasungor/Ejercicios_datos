```python
#3: Generar dataframes con los formatos pedidos
import pandas as pd
import numpy as np
df = pd.read_csv('Ejercicio 3 Guia 1 productos.csv')
df.head()
```

```python
#A: Promedio con formato | codigo producto | promedio de ventas por mes |
dfA = df.groupby('codigo producto').agg('mean')
dfA.head()
```

```python
dfA = dfA.drop(['codigo fabricante','mes'],axis = 1)
dfA.head()
```

```python
#B: Minimo, maximo y promedio de ventas con formato pedido en pdf
dfB = df.groupby('codigo producto').agg(min)
dfB = dfB.drop(['codigo fabricante','mes'],axis = 1)
dfB.columns = ['Minimo Ventas Mensual']
dfB['Maximo Ventas Mensual'] = df.groupby('codigo producto').agg(max)['ventas mensuales']
dfB['Promedio Ventas Mensual'] = df.groupby('codigo producto').agg('mean')['ventas mensuales']
dfB = dfB.stack()
dfB
```

```python
#C: Minimo, maximo y promedio con un formato unstacked
dfC = dfB.unstack()
dfC
```

```python
#D: Hacer un pivot de lo pedido en el punto C, como mostrado en el pdf
dfD = dfC.reset_index()
dfD
```

```python
dfD = dfD.pivot_table(columns = 'codigo producto')
dfD
```

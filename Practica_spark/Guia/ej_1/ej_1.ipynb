```python
from pyspark.sql import *
import pandas as pd
import math as mt
#import datetime as dt
#from pyspark.sql.functions import *
#from pyspark import SparkContext
#from pyspark.sql import SQLContext
```

```python
#El csv esta como padrón,materia,nota,fecha, no lo escribo porque sino lo lee spark igual

#Pruebo crear los contexts y leer como csv
spark = SparkSession.builder.getOrCreate()
sc = spark.sparkContext
sqlContext = SQLContext(sc)
rdd = sqlContext.read.csv("notas.csv")
#rdd.collect()

#No se que hacer con esto de que mete rows en vez de leer normalmente
```

```python
#Pruebo leer como text y generar registros que contengan cada dato
#padrón,materia,nota,fecha
rdd = sc.textFile("notas.csv")
rdd_regs_cambiados = rdd.map(lambda x: x.split(","))
rdd_regs_cambiados = rdd_regs_cambiados.map(lambda x: \
             (pd.to_numeric(x[0], "coerce"), x[1], pd.to_numeric(x[2], "coerce"), pd.to_datetime(x[3], "coerce")))
```

```python
#A: Cuántos alumnos aprobaron al menos 1 materia en los últimos 2 años
ultimo_anio = rdd_regs_cambiados.takeOrdered(1, lambda x: -x[3].year)[0][3].year
rdd_regs_cambiados.filter(lambda x:(x[3].year >= ultimo_anio - 1)and(x[2]>=4)).\
            map(lambda x:x[0]).distinct().count()
```

```python
#B: Un RDD conteniendo el promedio de notas de cada alumno de la forma (padrón, promedio)
promedios = rdd_regs_cambiados.map(lambda x:(x[0],(x[2], 1))).reduceByKey(lambda x,y:(x[0]+y[0], x[1]+y[1])).\
                    map(lambda x:(x[0], x[1][0]/x[1][1]))
promedios.collect()
```

```python
#C:El nombre y apellido del alumno con mejor promedio. 
#Para esto puede utilizarse un segundo RDD alumnos con registros (padron, nombre y apellido)
nombres = sc.textFile("nombres.csv")
nombres = nombres.map(lambda x:x.split(",")).map(lambda x:(pd.to_numeric(x[0], "coerce"), x[1]))
max_promedio = promedios.reduce(lambda x,y: x if x[1]>y[1] else y)[1]
mejores_notas = promedios.filter(lambda x: mt.isclose(max_promedio, x[1], abs_tol=0.005))
mejores_notas.leftOuterJoin(nombres).collect()
```

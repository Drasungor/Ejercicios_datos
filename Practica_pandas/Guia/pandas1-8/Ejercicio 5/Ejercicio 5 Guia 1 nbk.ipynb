```python
dfA = df.loc[df['mes_nacimiento'] == 10]
dfA = dfA.loc[df['anio_nacimiento'] == 2017]
dfA.head()
```

```python
dfA['counter'] = 1
octubre2017 = dfA.groupby('id_hospital').agg('sum')
octubre2017.head()
```

```python
dfH = dfH.set_index(dfH['id_hospital'])
dfH['promedio octubre 2017'] = octubre2017['counter']
dfH.head()
```

```python
dfH = dfH.fillna(0)
dfH.head()
```

```python
dfResultado = dfH.loc[dfH['promedio_nacimientos_mensual'] < dfH['promedio octubre 2017']]
dfResultado.head()
```

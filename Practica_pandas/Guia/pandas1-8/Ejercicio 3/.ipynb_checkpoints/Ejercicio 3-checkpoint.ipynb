```python
import pandas as pd
import numpy as np

productos = pd.read_csv('Ejercicio 3 Guia 1 productos.csv')
productos.head()
```

## Ejercicio a

```python
promedio_productos = productos.groupby('codigo producto').agg({'ventas mensuales':'mean'})
promedio_productos.rename(columns={'ventas mensuales':'promedio_de_ventas'}, inplace=True)
promedio_productos.sample(5)
```

## Ejercicio b

```python
ventas_productos = productos.groupby('codigo producto').agg({'ventas mensuales':['mean','max','min']})
ventas_productos.rename(columns={'mean':'promedio', 'max':'cantidad maxima', 'min':'cantidad minima'}, inplace=True)
ventas_productos.head()
```

```python
nivel0 = ventas_productos.columns.get_level_values(0)
nivel1 = ventas_productos.columns.get_level_values(1)
ventas_productos.columns = nivel1 + " de " + nivel0
ventas_productos.head()
```

```python
ventas_productos_final = ventas_productos.stack()
ventas_productos_final.head(10)
```
